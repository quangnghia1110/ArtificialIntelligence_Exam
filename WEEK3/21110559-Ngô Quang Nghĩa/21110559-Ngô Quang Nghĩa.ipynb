```python
# BÃ i 6
import pandas as pd
data = pd.read_csv('https://drive.google.com/uc?id=1C_Zf88mLfOn4AnNHVYttlgUUCcqIOlo4&export=download')
print(data.describe())
```

```python
# BÃ i 7
import matplotlib.pyplot as plt
vic_count = data.groupby('dog_type')['vic_name'].count()
print(vic_count)
plt.pie(vic_count.values, labels=vic_count.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
# BÃ i 8
class Attack_case:
    def __init__(self,date,year,city,state,vic_name,vic_age,dog_type,desc):
        self.date = date
        self.year = year
        self.city = city
        self.state = state
        self.vic_name = vic_name
        self.vic_age = vic_age
        self.dog_type = dog_type
        self.desc = desc
list_attack_case = []
for i in range(len(data)):
    row = data.iloc[i]
    attack_case = Attack_case(row['date'],row['year'],row['city'],row['state'],row['vic_name'],row['vic_age'],row['dog_type'],row['desc'])
    list_attack_case.append(attack_case)
for i in range(len(list_attack_case)):
    attack_case = list_attack_case[i]
    print("----------------")
    print(f"Case {i+1}:")
    print(f"\tDate: {attack_case.date}")
    print(f"\tYear: {attack_case.year}")
    print(f"\tCity: {attack_case.city}")
    print(f"\tState: {attack_case.state}")
    print(f"\tVic_name: {attack_case.vic_name}")
    print(f"\tVic_age: {attack_case.vic_age}")
    print(f"\tDog_type: {attack_case.dog_type}")
    print(f"\tDesciption: {attack_case.desc}")
```
